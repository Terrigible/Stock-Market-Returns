```python
import os
import time

from selenium.webdriver import Chrome, ChromeOptions
```

```python
options = ChromeOptions()
```

```python
prefs = {
    "download.default_directory": os.path.join(os.getcwd(), "data"),
    "download.prompt_for_download": False,
}
options.add_experimental_option("prefs", prefs)
options.add_argument("--headless=new")
```

```python
driver = Chrome(options=options)
```

```python
os.remove(os.path.join(os.getcwd(), "data", "ie_data.xls"))
```

```python
driver.get("https://shillerdata.com/")
```

```python
driver.find_element(
    "xpath", "/html/body/div[1]/div/div/div[2]/div/div/section/div/div[2]/div[1]/div/a"
).click()
```

```python
for i in range(50):
    time.sleep(0.1)
    if os.path.exists(os.path.join(os.getcwd(), "data", "ie_data.xls")):
        break
else:
    raise RuntimeError("Download failed")
```

```python
driver.quit()
```
